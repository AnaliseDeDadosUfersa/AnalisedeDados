O MultinomialNB é uma das duas variantes clássicas ingênuas de Bayes usadas na classificação de texto, onde os dados são tipicamente representados como contagens vetoriais de palavras. O ComplementNB é uma adaptação do algoritmo padrão MultinomialNB que é particularmente adequado para conjuntos de dados desequilibrados. Especificamente, o ComplementNB usa estatísticas do complemento de cada classe para calcular os pesos do modelo [Zules Acosta 2019].

</p>
<p>
GaussianNB is a special type of naive bayes algorithm, used specifically when resources have continuous values. It is also assumed that all features are following a Gaussian distribution, that is, normal distribution. Meanwhile, BernoulliNB is used for data distributed according to Bernoulli's multivariate distributions; that is, there may be several features, but each is assumed to be a binary (boolean) value variable [Zules Acosta 2019].
</p>
<p>
MultinomialNB is one of two naive classic variants of Bayes used in text classification, where data is typically represented as vector word counts. ComplementNB is an adaptation of the standard MultinomialNB algorithm that is particularly suitable for unbalanced data sets. Specifically, ComplementNB uses complement statistics for each class to calculate the weights of the [Zules Acosta 2019] model.
</p>
