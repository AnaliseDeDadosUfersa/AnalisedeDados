```python
#### import pandas as pd
from functools import reduce
import math as mt
import matplotlib.pyplot as plot
import pandas as pd
%matplotlib inline
class DataBase:
    def __init__ (self, link):
        try:
            self.data = pd.read_csv(link)                 #Aqui é onde esta os dados principais
            self.entropia = 0                             #Aqui estará a entropia do target
            self.header = list(self.data.keys())          #Aqui estará a lista de cabeçarios
            self.Translation = []                         #Aqui estará a lista de dicionario de traduçoes
            for i in range(len(self.header)):
                self.Translation.append(list(self.data[self.header[i]].value_counts().keys())) #prenche a traducao com os elementos contidos na coluna
        except:
            print("Link is in trouble.")               #para qualquer erro causado ao carregar o arquivo o objeto não sera
                                                       #criado.
    def modifyString(self, palavra):
        #Esta função altera a string passada, retirando pontos, 
        #virgulas e barras e substituindo por espaços em brancos.
        temp = ""
        for i in str(palavra):
            if i.isalpha() or i.isnumeric() or i.isspace():
                temp += i
            else:
                temp += " "
        return temp
            
    def modifyData(self, dicionario, ind):
        #Este metodo recebe uma lista de dicionarios e altera os dados substituindo os dados presentes nele
        #pelos dados passados.
        
        #Utilizando um dicionario a função irá comparar se o dado de certo indice na coluna condiz com a chave do dicionario
        #em caso afirmativo o valor da chave sera atribuida a uma lista de valores no exato indice em que o dado indica.
        #O dicionario basicamente serve como um conversor.
        #Exemplo: 
        #    Supondo os seguintes dados em lista [maça, abacaxi, banana]. 
        #    O seguinte dicionario; {maça:maça-verde, Default: fruta}
        #    A saida da função será uma nova lista da seguinte ordem: [maça-verde, fruta, fruta] 
        
        #Os parametros estão relacionados aos Atributos filtros, basicamente se for encontrado um atributos Filtro dentro de 
        #dados sera atribuido o parametro do mesmo idice. O ultimo elemento do filtro é o elemento default, se todos falharem ele
        #sera atribuido.
        
        #Armazena os valores das chaves e as chaves em duas listas diferentes
        list_values = list(dicionario.values())
        keys        = list(dicionario.keys())
        
        for j in ind:                                   #Para cada indice passado, este indice se refere a coluna em numeros.
            for l in range(0, len(list_values)):        #Para cada lista de valores dentro da lista de valores
                for i in range(0, len(list_values[l])): #Para cada valor dentro da lista
                    for k in self.data[self.header[j]]: #Para cada valor dentro da lista de valores da coluna
                        if self.modifyString(list_values[l][i]).upper() in self.modifyString(k).upper():
                            self.data[self.header[j]] = self.data[self.header[j]].apply(lambda x: str(x).replace( str(k), keys[l]))
                if (l == len(list_values) - 1):
                    for k in self.data[self.header[j]]:
                        if k not in keys:
                            print(k)
                            self.data[self.header[j]] = self.data[self.header[j]].apply(lambda x: str(x).replace( str(k), keys[l]))
        #Altera a lista de tradução
        for i in ind:
            self.Translation[i] = self.convertDict(dicionario)
                        
    def modifyHeader(self, list_header, ind):
        #Este metódo ira alterar os cabeçarios mardos pelo indice ind.
        temp = list(self.data.columns)
        for i in range(0, len(ind)):
            temp.remove(a[i])
            temp.insert(i, list_header[i])
        #altera as colunas somente se o tamanho da nova lista for igual ao tamanho da lista anterior
        if len(temp) == len(self.data.columns):
            self.data.columns = temp
        else: 
            print("Warning: Problems in the past parameters!")
            
    def modifyHeader(self, list_header):
        #altera as colunas somente se o tamanho da nova lista for igual ao tamanho da lista anterior
        if len(list_header) == len(self.data.columns):
            self.data.columns = list_header
        else: 
            return("Warning: Problems in the past parameters!")
        self.header = list(self.data.keys())
    
    def dropColumn(self, ind):
        #Tira as colunas indicadas pela lista ind
        clone = self.data
        clonec = []
        clonec.extend(self.Translation)
        
        k = 0
        tamt = len(self.Translation)
        temp = 0
        for i in ind:
            clone = clone.drop(self.data.keys()[i], axis = 1)
            
            clonet = []
            
            #Verifica a ordem dos numeros para definir se deve-se subtrair ou não
            if temp > i:
                k = i
            else:
                k += i
            
            for j in range(len(clonec)):
                if j != k:
                    clonet.append(clonec[j])
            k = len(clonet) - tamt
            clonec = clonet
            temp = i
        msg = input("Do you want to save this operation? Y/N ")
        if(msg.upper() == "Y"):
            self.data = clone
            self.Translation = clonec
            self.header = list(self.data.keys())
        else: 
            print("No column changed")
            
    def dropLines(self, ind):
        #Tira as linhas indicadas pela lista ind
        clone = self.data
        for i in ind:
            clone = clone.drop(i, axis = 0)
        msg = input("Do you want to save this operation? Y/N ")
        if(msg.upper() == "Y"):
            self.data = clone
        else: 
            print("No rows changed")
    
    def showGraphic(self, i):
        fig, piet = plot.subplots(figsize=(8, 5), subplot_kw=dict(aspect="equal"))
        
        a, b, c = piet.pie(self.data[self.header[i]].value_counts().values, 
                           labels = self.data[self.header[i]].value_counts().index, 
                           autopct = '%1.1f%%', labeldistance = 1.2, pctdistance = 0.8)
        
        piet.legend(self.Translation[i], title="Legend", loc= 8, bbox_to_anchor=(0, -0.3, 1, 0), ncol = 3)
        plot.setp(c, size=8, weight="bold")
        piet.set_title(self.header[i])
        plot.show()
        
    def showGraphict(self, i, j):
        fig, piet = plot.subplots(2, figsize=(12, 10), subplot_kw=dict(aspect="equal"))
        
        a, b, c = piet[0].pie(self.data[self.header[i]].value_counts().sort_index().values, 
                           labels = self.data[self.header[i]].value_counts().sort_index().index, 
                           autopct = '%1.1f%%', labeldistance = 1.2, pctdistance = 0.8)
        
        d, f, g = piet[1].pie(self.data[self.header[j]].value_counts().sort_index().values, 
                           labels = self.data[self.header[j]].value_counts().sort_index().index, 
                           autopct = '%1.1f%%', labeldistance = 1.2, pctdistance = 0.8)
        
        piet[0].legend(self.Translation[i], title= (self.header[i]) ,  loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), ncol = 2)
        piet[1].legend(self.Translation[j], title= (self.header[j]),  loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), ncol = 2)
        
        plot.setp(c, size=8, weight="bold")
        plot.setp(g, size=8, weight="bold")
        piet[0].set_title(self.header[i])
        piet[1].set_title(self.header[j])
        
        plot.show()
            
            
    def saveCsv(self, nome):
        try:
            self.data.to_csv(nome , sep = ',', index=False, encoding = 'utf-8', quotechar = '"', line_terminator = '\r\n')
        except:
            print("Error in operation!!!")
            
    def relationshipTable(self, ncoluna):
        tabela = {}
        l1 = list(self.data[self.header[0]].values)
        l2 = list(self.data[self.header[ncoluna]].values)
        
        qtl1 = list(self.data[self.header[0]].value_counts().keys())
        qtl2 = list(self.data[self.header[ncoluna]].value_counts().keys())
        
        ls = []
        
        for j in qtl1:
            ls = []
            for k in qtl2:
                temp = 0
                for i in range(0, len(l2)):
                    if l1[i] == j and l2[i] == k:
                        temp += 1
                ls.append(temp)
            tabela[j] = ls
            
        return tabela
    
    def relationshipTablein(self, xcoluna, ncoluna):
        tabela = {}
        l1 = list(self.data[self.header[xcoluna]].values)
        l2 = list(self.data[self.header[ncoluna]].values)
        
        qtl1 = list(self.data[self.header[xcoluna]].value_counts().keys())
        qtl2 = list(self.data[self.header[ncoluna]].value_counts().keys())
        
        ls = []
        
        for j in qtl1:
            ls = []
            for k in qtl2:
                temp = 0
                for i in range(0, len(l2)):
                    if l1[i] == j and l2[i] == k:
                        temp += 1
                ls.append(temp)
            tabela[j] = ls
            
        return tabela
    
    def gainInformation(self, ncoluna):
        #Rever esse calculo pois o valor deveria ser entre 0 e 1 nao esta batendo
        entropiaR = self.entropia
        valores = list(self.relationshipTable(ncoluna).values())
        soma = 0
        sm = []
        for i in valores:
            for j in i:
                soma += j
            
        for j in range(0, len(valores[0])):
            a = 0
            while(a < len(valores)):
                sm.append(valores[a][j])
                a += 1
            b = reduce(lambda x, y: x + y, sm)
            entropiaR -= (b / soma) * self.entropiaS(sm, soma)
            sm = []
        return entropiaR
    
    def setEntropia(self, ncoluna):
        #Define a entropia do sistema
        #Deve ser a entropia da coluna target
        self.entropia = self.entropiaS(self.data[self.header[ncoluna]].value_counts().values, len(self.data))
    
    def entropiaS(self, data, qt):
        #Rever esse calculo pois o valor deveria ser entre 0 e 1 nao esta batendo
        entropiaR = 0
        valores = data
        for j in valores:
            try:
                entropiaR -= (int(j) / qt) * mt.log2((int(j) / qt))
            except:
                entropiaR -= (int(j) / qt) * 0
        return entropiaR
    
    def convertDict(self, dicionario):
        listak = []
        for j in list(dicionario.values()):
            if len(j) > 1:
                listak.append(reduce(lambda x, y: x + " - " + y, j))
            else:
                listak.append(j[0])
        return listak

    
    def pdRelation(self, coluna, linha):
        dat = pd.DataFrame(self.relationshipTablein(coluna,linha))

        dat.index = self.Translation[linha]
        dat.columns = self.Translation[coluna]
        return dat
```

<h4>Nota:</h4>
<p>
Percebeu-se, a necessidade de algumas modificações nos dados. Sendo assim, a estratégia empregada foi definir filtros que são na realidade dicionários, que no método processarDados() tem os values() testados e substituídos pelas keys(). A vantagem dessa estratégia é que diferentemente do método apply() do pandas, ela não necessita ter exatamente a palavra que consta na coluna em que será testado, por exemplo, tendo na coluna a seguinte string: "Qual sua renda", se um dos dicionários tiver um valor "renda", com a key "1", toda a palavra "renda" será substituída por "1", esse mesmo dicionário poderá substituir mais de um valor por uma key, pois o método espera que dentro do dicionário sua key aponte para uma lista de valores, que podem ser substituído por essa key. Outra vantagem é a questão da possibilidade de poder ser passado mais de uma uma coluna por vez, a partir de uma lista, exemplo: suponha-se que se deseje aplicar o dicionário supracitado acima ({"1" : "renda"}) em várias colunas, no lugar de chamar várias vezes a mesma função, pode-se passar apenas uma lista de colunas para se aplicar o filtro. Por fim, o funcionamento básico dessa estratégia é a chamada do método <b>processarDados(filtro, lista_colunas).</b></p>

<br/>

<h4> Note: </h4>
<p>
It was noticed, the need for some modifications in the data. Therefore, the strategy employed was to define filters that are actually dictionaries, which in the processData () method have the values() tested and replaced by the keys (). The advantage of this strategy is that unlike the apply () method of pandas, it does not need to have exactly the word that appears in the column in which it will be tested, for example, having the following string in the column: "What is your income" if one of the dictionaries have an "income" value, with the key "1", the whole word "income" will be replaced by "1", that same dictionary can replace more than one value with a key, since the method expects that in the dictionary its key point to a list of values, which can be overridden by that key. Another advantage is the possibility of being able to be passed more than one column at a time, from a list, for example: suppose you want to apply the dictionary above ({"1": "income"}) in several columns, instead of calling the same function several times, you can only pass a list of columns to apply the filter. Finally, the basic function of this strategy is to call the <b> processData method (filter, list_columns). </b>
</p>

## 4.2 - Normalization

<p>
A normalização lida com a mudança e padronização da dimensão de escalas dos dados, assim sendo, atributos podem ser normalizados com casas decimais deliberadas após a vírgula ou mecanismos que definem uma
única escala para todos os atributos (e.g., de 0 a 1). Em ambos os casos, a normalização pode aumentar o desempenho dos modelos, posto que permite ao algoritmo executar de forma mais simples omitindo cálculos extensos.
</p>

<br/>

<p>
Normalization deals with the change and standardization of the dimension of scales of the data, so, attributes can be normalized with decimal deciles after the comma or mechanisms that define a

### Model bayes

<p>
Em seguida, tratou-se de ajustar as variáveis resultantes da atividade anterior no algoritmo escolhido. Salienta-se que o algoritmo optado foi o naive bayes. Percebeu-se que as bibliotecas do python, oferecem 4 (quatro) variações do naive bayes (GaussianNB, BernoulliNB, MultinomialNB e ComplementNB). 
</p>
<p>
O GaussianNB é um tipo especial de algoritmo naive bayes, empregado especificamente quando os recursos possuem valores contínuos. Também é se presume que todos os recursos estão seguindo uma distribuição gaussiana, isto é, distribuição normal. Enquanto isso, o BernoulliNB é utilizado para dados distribuídos de acordo com as distribuições multivariadas de Bernoulli; isto é, pode haver vários recursos, mas cada um é assumido como sendo uma variável de valor binário (booleano) [Zules Acosta 2019].
</p>
<p>
O MultinomialNB é uma das duas variantes clássicas ingênuas de Bayes usadas na classificação de texto, onde os dados são tipicamente representados como contagens vetoriais de palavras. O ComplementNB é uma adaptação do algoritmo padrão MultinomialNB que é particularmente adequado para conjuntos de dados desequilibrados. Especificamente, o ComplementNB usa estatísticas do complemento de cada classe para calcular os pesos do modelo [Zules Acosta 2019].
</p>

<br/>


<p>
Next, we tried to adjust the variables resulting from the previous activity in the chosen algorithm. It should be noted that the algorithm chosen was the naive bayes. It was found that the python libraries, offer four (4) variations of naive bayes (GaussianNB, BernoulliNB, MultinomialNB and ComplementNB).
</p>
<p>
GaussianNB is a special type of naive bayes algorithm, used specifically when resources have continuous values. It is also assumed that all features are following a Gaussian distribution, that is, normal distribution. Meanwhile, BernoulliNB is used for data distributed according to Bernoulli's multivariate distributions; that is, there may be several features, but each is assumed to be a binary (boolean) value variable [Zules Acosta 2019].
</p>
<p>
MultinomialNB is one of two naive classic variants of Bayes used in text classification, where data is typically represented as vector word counts. ComplementNB is an adaptation of the standard MultinomialNB algorithm that is particularly suitable for unbalanced data sets. Specifically, ComplementNB uses complement statistics for each class to calculate the weights of the [Zules Acosta 2019] model.
</p>

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.model_selection import train_test_split #para deixar o teste e o treinamento aleatorio
from sklearn.metrics import accuracy_score
class modelNaive():
        def __init__(self, link):
            self.data = DataBase(link)
            self.vect = []             #Contem as colunas que não são target
            self.target = []           #Contem a coluna target em forma de vetor
            self.totalVet = []
            self.model = GaussianNB() #modelo default
            
            #variaveis de treinamento
            self.x_train = []
            self.x_test  = []
            self.y_train = [] 
            self.y_test  = []
            
            for i in range(0, len(self.data.data)):
                self.totalVet.append(list(self.data.data.iloc[i].values))
        
        def defineTarget(self, ind):
            self.vect = []
            self.target = []
            temp = []
            for i in self.totalVet:
                self.target.append(int(i[ind]))
                for j in range(len(i)):
                    if (j != ind):
                        temp.append(int(i[j]))
                self.vect.append(temp)
                temp = []
                
        def defineModel(self, choice):
            #instancia um dos 4 modelos de naive bayes definidos pelo sklearn
            if(choice == 1):
                self.model = GaussianNB()
            elif(choice == 2):
                self.model = BernoulliNB(binarize = True)
            elif(choice == 3):
                self.model = MultinomialNB()
            elif(choice == 4):
                self.model = ComplementNB()
            else:
                print("Past value does not match any of the 4 possible values !!!")
        
        def allModel(self):
            #instancia os 4 modelos de naive bayes definidos pelo sklearn
            model1 = GaussianNB()
            model2 = BernoulliNB(binarize = True)
            model3 = MultinomialNB()
            model4 = ComplementNB()
            self.model = [model1, model2, model3, model4]
            
        def trainModel(self, pctTestSize):
            try:
                self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(self.vect, self.target, test_size = pctTestSize, random_state = 2) #define o treino e o teste   
            except:
                print("Ponse invalidate, please enter a number between 0 and 1.")
            if(len(self.model) == 4):
                #treino do modelo
                    for i in range(len(self.model)):
                        self.model[i].fit(self.x_train, self.y_train)
                        print(self.model[i])
            else:
                self.model.fit(self.x_train, self.y_train)
                
        def errorModel(self):
            if (len(self.model) == 4):
                for i in range(len(self.model)):
                    print(type(self.model[i]))
                    print("Errors : ", (self.model[i].predict(self.x_test) != self.y_test).sum())
            else:
                print(type(self.model))
                print("Errors : ", (self.model.predict(self.x_test) != self.y_test).sum()) 
                
        def errorModelInd(self, begin, end):
                if(len(self.model) == 4):
                    for i in range(len(self.model)):
                        print("Errors in model",i," :", (self.model[i].predict(self.vect[begin:end]) != self.target[begin:end]).sum())
                else:
                    print("Errors in model :", (self.model.predict(self.vect[begin:end]) != self.target[begin:end]).sum())
                
        def compareModel(self, begin, end):
            try:
                print("           Estimate of result for each algorithm of naive bayes")
                print("=====================================================================================")
                print("N°     GaussianNB" , "     ", "BernoulliNB", "     ", "MultinomialNB", "     ", 
                      "ComplementNB", "  ", "target" )
                for i in range(begin , end):
                    print(str(i).zfill(2),"       ", self.model[0].predict([self.x_test[i]]) , "             " , 
                          self.model[1].predict([self.x_test[i]]), "              ",self.model[2].predict([self.x_test[i]]), "               ",
                          self.model[3].predict([self.x_test[i]]),"        " ,self.y_test[i])
                print("=====================================================================================")
                print("ERRORS:     ", (self.model[0].predict(self.x_test[begin:end]) != self.y_test[begin:end]).sum(), "               ",
                      (self.model[1].predict(self.x_test[begin:end]) != self.y_test[begin:end]).sum(), "                " ,
                      (self.model[2].predict(self.x_test[begin:end]) != self.y_test[begin:end]).sum(),"                 ", 
                      (self.model[3].predict(self.x_test[begin:end]) != self.y_test[begin:end]).sum())
                print("* Test Vector Size: ", len(self.x_test[begin:end]), "positions.")
            except  ValueError:
                print("Past value is inconsistent with the required type, please pass a whole number !!!")
            except IndexError:
                print("The last range does not belong to the test list, please enter a value between 0 and", len(self.x_test))
                
        def probabilitySucessModel(self):
            if (len(self.model) == 4):
                for i in range(len(self.model)):
                    print("Probability of model correctness",i," : ", self.model[i].score(self.x_test, self.y_test) * 100, "%")
            else:
                print("Probability of model correctness:", model.score(self.x_test, self.y_test) * 100, "%")
            
        def probabilitySucessModelInd(self, begin, end):
            if(len(self.model) == 4):
                try:
                    for i in range(len(self.model)):
                        print("Chance of model hit",i,":", self.model[i].score(self.vect[begin:end], self.target[begin:end])*100, "%.")
                except IndexError:
                    print("The last range does not belong to the test list, please enter a value between 0 and ", len(self.x_test))
                except ValueError:
                    print("The past interval has inconsistencies or does not exist in the list !!!")
            else:
                try:
                    print("Chance of model hit: ", self.model.score(self.vect[begin:end], self.target[begin:end])*100, "%.")
                except IndexError:
                    print("The last range does not belong to the test list, please enter a value between 0 and ", len(self.x_test))
                except ValueError:
                    print("The past interval has inconsistencies or does not exist in the list !!!")
```
