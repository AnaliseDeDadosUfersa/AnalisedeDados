```python
#Agregação

years     ={'0' : ["2006 - 2007"],
            '1' : ["2008 - 2009"], 
            '2' : ["2010 - 2011"], 
            '3' : ["2012 - 2013"],
            '4' : ["2014 - 2015"], 
            '5' : ["2016 - 2017"],
            '6' : ["2018 - 2019"]} #> 2019 é o default

cities = {  '0' : ["Mossoró"],
            '1' : ['Cities of the RN'], 
            '2' : ['Cities of the CE'],
            '3' : ['Others']} #outros é o default

income =   {'0' : ['Less than 500'],
            '1' : ['500  -| 1500'], 
            '2' : ['1500 -| 3000'], 
            '3' : ['More than 3000']} #> 3000 é o default

ethnicity ={'0' : ['White'], 
            '1' : ["Brown or Black"], 
            '2' : ['Others']} #outros é o default


schoolingParents = {'0' : ["fundamental level"],
                    '1' : ["medium level"],
                    '2' : ["higher level"],
                    '3' : ['Others']} #outros é o default

#Normalizaçãp
YesNo  = { '0' : ["Yes"], "1" : ["No"]}

situacion =   {'0' : ["Concluding"], 
                '1': ["Evaded"]} #Desistente é o default

school  =    { '0' : ['Private'], "1" : ["Public"]}#publica é o default

typeSchool = { '0' : ['Private'],
               '1' : ['Federal'],
               '2' : ['Public']} #privada é o default

yields =    {'0' : ['Approved in all'], 
             '1' : ['failed in up to two subjects'],
             '2' : ['failed in up to five subjects'],
             '3' : ["More than five disciplines"]} #Mais de cinco é o default
```

```python
dictionary = [years, YesNo, school, typeSchool, cities, income, ethnicity, yields, schoolingParents]
freq = [[1,2], [6,7,10,12,13,14,15,16,17,18,19], [3], [4], [5], [8], [9], [11], [20,21]]
```

```python
dict1 = {}
for i in range(len(dictionary)):
    for j in freq[i]:
        x = j - 1
        dict1[modelo.data.data.columns[x]] = modelo.characteristicAtt(x, list(dictionary[i].values()))
```

```python
abc = pd.DataFrame()
```

```python
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
for i in range(len(dict1)):
    a.append(dict1[modelo.data.data.columns[i]][0][0])
    b.append(dict1[modelo.data.data.columns[i]][1][0])
    c.append(dict1[modelo.data.data.columns[i]][0][1])
    d.append(dict1[modelo.data.data.columns[i]][1][1])
    e.append(dict1[modelo.data.data.columns[i]][0][2])
    f.append(dict1[modelo.data.data.columns[i]][1][2])
    g.append(dict1[modelo.data.data.columns[i]][0][3])
    h.append(dict1[modelo.data.data.columns[i]][1][3])
abc['Concluding Characteristic GaussianNB'] = a
abc['Character of the evaded GaussianNB'] = b
abc['Concluding Characteristic BernoulliNB'] = c
abc['Character of the evaded BernoulliNB'] = d
abc['Concluding Characteristic MultinomialNB'] = e
abc['Character of the evaded MultinomialNB'] = f
abc['Concluding Characteristic ComplementNB'] = g
abc['Character of the evaded ComplementNB'] = h
```

```python
abc.index = dict1.keys()
```
