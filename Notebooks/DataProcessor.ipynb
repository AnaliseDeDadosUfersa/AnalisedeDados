<h4>Nota:</h4>
<p>
Percebeu-se, a necessidade de algumas modificações nos dados. Sendo assim, a estratégia empregada foi definir filtros que são na realidade dicionários, que no método processarDados() tem os values() testados e substituídos pelas keys(). A vantagem dessa estratégia é que diferentemente do método apply() do pandas, ela não necessita ter exatamente a palavra que consta na coluna em que será testado, por exemplo, tendo na coluna a seguinte string: "Qual sua renda", se um dos dicionários tiver um valor "renda", com a key "1", toda a palavra "renda" será substituída por "1", esse mesmo dicionário poderá substituir mais de um valor por uma key, pois o método espera que dentro do dicionário sua key aponte para uma lista de valores, que podem ser substituído por essa key. Outra vantagem é a questão da possibilidade de poder ser passado mais de uma uma coluna por vez, a partir de uma lista, exemplo: suponha-se que se deseje aplicar o dicionário supracitado acima ({"1" : "renda"}) em várias colunas, no lugar de chamar várias vezes a mesma função, pode-se passar apenas uma lista de colunas para se aplicar o filtro. Por fim, o funcionamento básico dessa estratégia é a chamada do método <b>processarDados(filtro, lista_colunas).</b></p>

<br/>

<h4> Note: </h4>
<p>
It was noticed, the need for some modifications in the data. Therefore, the strategy employed was to define filters that are actually dictionaries, which in the processData () method have the values() tested and replaced by the keys (). The advantage of this strategy is that unlike the apply () method of pandas, it does not need to have exactly the word that appears in the column in which it will be tested, for example, having the following string in the column: "What is your income" if one of the dictionaries have an "income" value, with the key "1", the whole word "income" will be replaced by "1", that same dictionary can replace more than one value with a key, since the method expects that in the dictionary its key point to a list of values, which can be overridden by that key. Another advantage is the possibility of being able to be passed more than one column at a time, from a list, for example: suppose you want to apply the dictionary above ({"1": "income"}) in several columns, instead of calling the same function several times, you can only pass a list of columns to apply the filter. Finally, the basic function of this strategy is to call the <b> processData method (filter, list_columns). </b>
</p>
